* [사이킷런 데이터셋 페이지] (https://scikit-learn.org/stable/modules/classes.html?highlight=datasets#module-sklearn.datasets)

* 데이터셋: 특정한 작업을 위해 데이터를 관련성있게 모아놓은 것

Iris DataSet: '붓꽃 150송이의 꽃받침, 꽃잎의 길이, 너비 데이터를 모아놓은 데이터셋'을 의미

```python
from sklearn.datasets import load_iris
```

```python
# iris와 관련된 데이터를 모두 불러옴.
iris = load_iris()
```

```python
# iris와 관련된 데이터가 모두 출력됨.
iris
```

```python
# iris와 관련된 전체적인 설명을 확인.
print(iris['DESCR'])
```

    :Attribute Information:
        - sepal length in cm : 꽃받침의 길이
        - sepal width in cm : 꽃받침의 너비
        - petal length in cm : 꽃잎의 길이
        - petal width in cm : 꽃잎의 너비
        - class:
                - Iris-Setosa
                - Iris-Versicolour
                - Iris-Virginica

```python
# iris의 데이터를 모두 가져옴.
data = iris['data']
data # 150개의 행과 4개의 종속변수를 출력함.
# type: ndarray
```

```python
target = iris['target']
target # '정답'과 유사한 개념
```

```python
# ndarray인 data를 dataFrame으로 변경
# 먼저 column을 지정
feature_names = iris['feature_names']
feature_names
```

```python
import pandas as pd
```

```python
# ndarray인 data를 dataFrame으로 변경
df_iris = pd.DataFrame(data, columns = feature_names)
df_iris.head() # 너무 길어서 head() 붙임
```

```python
# 오른쪽에 target(정답)을 붙여주고 싶은 경우
```

```python
df_iris['target'] = target
df_iris
```

```python
from sklearn.model_selection import train_test_split
```

```python
# train_test_split(독립변수, 종속변수, 테스트데이터의사이즈[기본값=0.25%. 75:25로 나뉘어진다는 의미], 시드값[random state. 아무값이나 상관없음.])
#  random_state 값을 동일하게 유지한다면 코드를 여러 번 실행할 때 항상 같은 데이터 분할 결과가 생성됨.

X_train, X_test, y_train, y_test = train_test_split(df_iris.drop('target',1), df_iris['target'], test_size=0.2, random_state=10)
```

```python
X_train.shape, X_test.shape # 120,4와 30,4로 8대2로 잘 섞임.
```

```python
y_train.shape, y_test.shape # 120, 30으로 8대2로 잘 섞임.
```

```python
X_train
```

```python
# 머신러닝 모델에 넣어 학습을 시켜볼 예정.
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
svc = SVC()
```

```python
svc.fit(X_train, y_train)
```

```python
# 시험을 봄
y_pred = svc.predict(X_test)
```

```python
print('정답률: ', accuracy_score(y_test, y_pred))
```
